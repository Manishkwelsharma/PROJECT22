```python
df.drop(['ID','Delivery_person_ID','Order_Date'],axis=1,inplace=True)
```

```python
def unique_values(df):
    for i in df.columns:
        print(i,end='\n\t')
        print(df[i].unique())
unique_values(df)
```

```python
obj=['Weather_conditions','Road_traffic_density','Festival','City','multiple_deliveries']
for i in obj:
    df[i].fillna(value=df[i].mode()[0],inplace=True)
```

```python
k=['Delivery_person_Age','Delivery_person_Ratings']
for i in k:
    m=df[i].mean()
    if i=='Delivery_person_Ratings':
        df[i].fillna(value=round(m,1),inplace=True)
    else:
        df[i].fillna(value=int(m),inplace=True)
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
def create_dummy(df):
    dummy_cols=['Weather_conditions','Road_traffic_density','Type_of_order','Type_of_vehicle','Festival','City']
    for i in dummy_cols:
        temp=pd.get_dummies(df[i],prefix=i,drop_first=True)
        df.drop(i,axis=1,inplace=True)
        df=pd.concat([df,temp],axis=1)
    return df
df=create_dummy(df)     
```

```python
df.info()
```

```python
from datetime import datetime, time
df['Time_Orderd'] = df['Time_Orderd'].astype(str)
df['Time_Order_picked'] = df['Time_Order_picked'].astype(str)
def time_diff(row):
    try:
        t1 = datetime.strptime(row['Time_Orderd'], '%H:%M').time()
        t2 = datetime.strptime(row['Time_Order_picked'], '%H:%M').time()
        time_diff = datetime.combine(datetime.today(), t2) - datetime.combine(datetime.today(), t1)
        return time_diff.seconds // 60
    except ValueError:
        return None
df['time_diff'] = df.apply(time_diff, axis=1)
df.drop(['Time_Orderd','Time_Order_picked'],axis=1,inplace=True)
```

```python
for i in df.columns[0:9]:
    try:
        df[i]=df[i].astype(int)
    except:
        print("error in",i)
```

```python
df.dropna(subset=['time_diff'],inplace=True)
```

```python
x=df.drop(['Time_taken (min)'],axis=1)
y=df['Time_taken (min)']
```

```python
from sklearn.preprocessing import MinMaxScaler
mnmx=MinMaxScaler()
x=mnmx.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
from sklearn.ensemble import RandomForestRegressor
lr=RandomForestRegressor()
lr.fit(x_train,y_train)
y_predlr=lr.predict(x_test)
```

```python
y_predlr=[round(i) for i in y_predlr]
```

```python
y_predlr[0:5]
```

```python
y_test[0:5]
```

```python
print(lr.score(x_train,y_train))
```

```python
print(lr.score(x_test,y_test))
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("r2_score is:",r2_score(y_test,y_predlr))
print("MAE is:",mean_absolute_error(y_test,y_predlr))
print("MSE is:",mean_squared_error(y_test,y_predlr))
print("RMSE is:",mean_squared_error(y_test,y_predlr,squared=False))
```
